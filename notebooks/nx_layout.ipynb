# graphviz_layout from networkX

```python
import requests
```

```python
url = 'http://localhost'
headers = {'Content-type': 'application/json'}
# parameters: parameters of graphviz_layout()
parameters = {'prog' : 'circo', 'root' : 0}
```

```python
data = open('../services/nx_layout/sample-data/sample.cx')
str_param = '?'
for key, value in parameters.items():
    str_param += str(key) + "=" + str(value) + '&'
    
r = requests.post(url=url+str_param, headers=headers, data=data)
print(parameters)
print('==')
r.json()
```
